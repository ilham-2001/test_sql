```python
import pandas as pd
```

### Product Info

```python
df = pd.read_csv('data/Data Bangkit - Info.csv')
df['cogs'] = df['cogs'].fillna(0)
df
```

```python
df.info()
```

```python
def formatAmountCogs(data):
  if type(data) is str:
    return int("".join(data.split(".")[0].split(",")))
  return data

print(formatAmountCogs("45,925.00"))
```

```python
df['cogs'] = df['cogs'].map(formatAmountCogs)
```

```python
df.to_csv("data/Data Bangkit - Info_clean.csv", index=False)
```

### Product Discount

```python
df = pd.read_csv("data/Data Bangkit - discount.csv")
df
```

```python
df.info()
```

```python
df["discount_pctg"] = df["discount_pctg"].map(lambda data: int(data.split("%")[0]))
df
```

```python
df.to_csv("data/Data Bangkit - discount_clean.csv", index=False)
```
